```python
#넷플릭스 추천 프로그램

import pandas as pd
data=pd.read_csv(r'C:\Users\h\hmkd1\m1_programming\netflix_titles_nov_2019.csv')
#data=pd.read_csv(r'C:\Users\fillo\OneDrive\바탕 화면\hmkd\netflix_titles_nov_2019.csv')

sdata=[]  #전체 데이터

for i in range(12):    #읽어온 데이터프레임 값을 리스트로 저장
    label=[]    #한 작품의 정보들 순서대로 저장 (쇼아이디, 제목, 감독, 배우...등)
    for row_index, row in data.iterrows():    #iterrows: 각 열의 인덱스와 값 출력
        label.append(row.iloc[i])      #iloc: 데이터프레임의 행이나 컬럼에 인덱스 값으로 접근
    sdata.append(label)     #모든 작품 정보들을 저장
    
rating=['G','TV-Y','TV-G','PG','TV-Y7','TV-Y7-FV','TV-PG','PG-13','TV-14','R','NC-17','TV-MA','NR']
rating_k=['전체관람가','영유아를 위한 프로그램','모든 연령이 시청할 수 있는 프로그램','부모지도하 전체관람가','7세 이상의 어린이를 위한 프로그램','7세 이상의 어린이를 위한 프로그램(more Fantasy Violence)','어린이가 시청하려면 보호자의 지도가 필요한 프로그램','13세 이상 관람가, 단 부모동반시 관람가능','14세 이상 관람가','17세 이상 관람가, 단 부모동반시 관람가능','청소년 관람불가','17세 이상 관람가','평가되지 않음']
for i in range(len(sdata[7])):     #등급 표기 -> 보기 편하게 한글로 수정
    if sdata[7][i] in rating:
        n=rating.index(sdata[7][i])
        sdata[7][i]=rating_k[n]

sdata[2]=list(map(str, sdata[2]))      #속성오류(str인데 float로 되어 있음) -> 타입 재지정

for i in range(len(sdata[2])):     #감독 'nan' -> '정보 없음'으로 변경
    if sdata[2][i]=='nan':
        sdata[2][i]='정보없음'

star_data=[0]*5837  #컨텐츠 평점 0으로 초기화
star_count=[0]*5837    #평점 매긴 사람 수
sd=['-']*5837
```

```python
print("==================================================================")
print("[안녕하세요, 넷플릭스 컨텐츠 룸((Netflix Contents Room)입니다]")

while True:
    print("==================================================================")
    service=input("<1. 검색>   <2. 컨텐츠 추천>   <3. 종료> \n[원하시는 서비스를 선택해주세요] => ")
    #############################################################
    #검색 서비스 (1)
    if service=='1' or service=='검색':
        print("==================================================================")
        print("[넷플릭스(Netflix) 컨텐츠 검색 서비스입니다]")
        title=input("[찾으려는 컨텐츠의 제목을 입력해주세요] => ")
        
        b_imsi=1
        while b_imsi==1:
            title=title.lower()     #소문자로 치환
            s_data=[]
            for i in range(len(sdata[1])):     #컨텐츠 제목들도 소문자로
                s_data.append(sdata[1][i].lower())
                s_data[i].replace(' ','')   #띄어쓰기 없애기
            title_list=[x for x in range(len(s_data)) if title in s_data[x]]   #검색한 문자열 포함하는 제목들의 위치값 전부
            
            print("\n[검색 결과입니다]\n■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■\n")
            c_imsi=1
            for i in range(len(title_list)):
                print(f"{c_imsi}. {sdata[1][title_list[i]]} ({sdata[6][title_list[i]]})")
                c_imsi+=1    #그냥 목록 예쁘라고 1.000, 2.000 이런 식으로 써 놓은거
            print("\n■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■")
                
            imsi=int(input("[다음 중 원하는 결과가 있다면 해당하는 숫자를, 없다면 '0'을 입력해주세요] => "))

            a_imsi=1
            while a_imsi==1:
                if 0 < imsi <= len(title_list):    #번호 선택
                    print(f"""\n[선택하신 컨텐츠의 정보입니다]  
■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■                  

평점: {sd[title_list[imsi-1]][:3]} ({star_count[title_list[imsi-1]]})
제목: {sdata[1][title_list[imsi-1]]} (감독: {sdata[2][title_list[imsi-1]]})
개봉 년도: {sdata[6][title_list[imsi-1]]}
출연진: {sdata[3][title_list[imsi-1]]}

구분: {sdata[11][title_list[imsi-1]]}
장르: {sdata[9][title_list[imsi-1]]}

국가: {sdata[4][title_list[imsi-1]]}
관람등급: {sdata[7][title_list[imsi-1]]}
러닝 타임(시즌): {sdata[8][title_list[imsi-1]]}

작품 설명: {sdata[10][title_list[imsi-1]]} 

■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■""")
                    star=input("[해당 작품에 평점을 매기겠습니까?(Y/N)]: ")
                    
                    if star=='y' or star=='Y':
                        star_data[title_list[imsi-1]]+=float(input("작품 평점을 입력하세요(/5): "))
                        star_count[title_list[imsi-1]]+=1
                        sd[title_list[imsi-1]]=str(star_data[title_list[imsi-1]]/star_count[title_list[imsi-1]])
                        print("컨텐츠에 평점이 저장되었습니다.")
                        
                    b_imsi-=1
                    a_imsi-=1
                
                elif imsi==0:    #결과에 없으면 재검색
                    title=input("\n[제목을 다시 입력해주세요] => ")
                    a_imsi-=1
            
                else:     #잘못된 입력
                    imsi=int(input("['종료(0)' 혹은 위의 숫자 중 하나를 선택해주세요]"))
            
    #############################################################
    #컨텐츠 추천 서비스 (2)
    elif service=='2' or service=='컨텐츠 추천 서비스' or service=='컨텐츠추천서비스':

        print("==================================================================")
        print("\n[넷플릭스(Netflix) 컨텐츠 맞춤 추천 서비스입니다]")
        print("\n[사용자의 국적, TV/Movie, 연령, 선호 장르 등에 따라 컨텐츠를 추천받을 수 있습니다]]")
        print("[다음의 물음에 답해주십시오. (질문을 패스하려면 '0'을 입력해주세요)]")
        
        n_data=[]
        list_data=[]
        list2_data=[]
        list3_data=[]
        list4_data=[]
        
        #1번 질문: 국가
        con=input("\n[사용자 국가를 입력해주세요] => ")
        a_imsi=1
        
        while a_imsi==1:
            if con=='0':
                print("['0'을 입력하셨습니다. 해당 질문을 패스합니다]")
                for i in range(len(sdata[4])):
                    list_data.append(i)
                a_imsi-=1

            else:
                con=con.lower()
                con=con.replace(' ','')

                sdata[4]=list(map(str, sdata[4]))      #속성오류 -> 타입 재지정
            
                for i in range(len(sdata[4])):     #컨텐츠 제목들도 소문자로
                    n_data.append(sdata[4][i].lower())
                    n_data[i]=n_data[i].replace(' ','')   #띄어쓰기 없애기
                    if con in n_data[i]:
                        list_data.append(i)         #list_data => 걸러낸 항목의 인덱스 주소 저장
            
                if not list_data:
                    con=input("[해당 국가의 컨텐츠가 존재하지 않습니다. 다시 입력해주세요]")
                else:
                    a_imsi-=1
        
        #2번 질문: tv or movie
        con=input("\n[TV프로그램(1) / 영화(2) 중 선호하는 항목을 선택해주세요] => ")
        a_imsi=1

        while a_imsi==1:
            if con=='0':
                print("['0'을 입력하셨습니다. 해당 질문을 패스합니다]")
                list2_data=list_data[:]   #list2_data => 2차로 걸러진 인덱스
                a_imsi-=1

            elif con=='1' or con=='TV프로그램' or con=='TV' or con=='tv프로그램' or con=='tv' :
                for i in list_data:
                    if sdata[11][i]=='TV Show': list2_data.append(i)
                a_imsi-=1

            elif con=='2' or con=='영화':
                for i in list_data:
                    if sdata[11][i]=='Movie': list2_data.append(i)
                a_imsi-=1
             
            else:
                con=input("[패스(0) / TV프로그램(1) / 영화(2) 중 하나를 선택해주세요] => ")
        

        #3번 질문: 미자 or 성인
        con=input("\n[만 17세 미만이신가요? Y/N (패스가 불가능한 질문입니다)] => ")
        a_imsi=1

        while a_imsi==1:
            if con=='Y' or con=='y':
                for i in list2_data:
                    if sdata[7][i] not in rating_k[10:]:   #청불 제외
                        list3_data.append(i)         #list3_data => 3번째 거르기
                a_imsi-=1

            elif con=='N' or con=='n':
                list3_data=list2_data[:]
                a_imsi-=1
        
            else:
                con=input("['Y' or 'N' 중 하나만 선택해주세요] => ")
        

        #4번 질문: 장르 선택
        con=input("\n[로맨스(1) / 코미디(2) / 다큐멘터리(3) / 판타지(4) / 공포_스릴러(5) 중 선호하는 장르를 선택해주세요] => ")
        a_imsi=1

        while a_imsi==1:
            if con=='0':
                print("['0'을 입력하셨습니다. 해당 질문을 패스합니다]")
                list4_data=list3_data[:]
                a_imsi-=1

            elif con=='1' or con=='로맨스':
                for i in list3_data:
                    if 'Romantic' in sdata[9][i]:
                        list4_data.append(i)
                a_imsi-=1

            elif con=='2' or con=='코미디':
                for i in list3_data:
                    if 'Comedies' in sdata[9][i]:
                        list4_data.append(i)
                a_imsi-=1

            elif con=='3' or con=='다큐멘터리' or con=='다큐':
                for i in list3_data:
                    if 'Documentaries' in sdata[9][i]:
                        list4_data.append(i)
                a_imsi-=1

            elif con=='4' or con=='판타지':
                for i in list3_data:
                    if 'Fantasy' in sdata[9][i]:
                        list4_data.append(i)
                a_imsi-=1

            elif con=='5' or con=='공포' or con=='스릴러' or con=='공포_스릴러':
                for i in list3_data:
                    if 'Horror' in sdata[9][i] or 'Thrillers' in sdata[9][i]:
                        list4_data.append(i)
                a_imsi-=1
            
            else:
                con=input("[패스(0) 로맨스(1) / 코미디(2) / 다큐멘터리(3) / 판타지(4) / 공포_스릴러(5) 중 하나를 선택해주세요] => ")
    
        print("\n[모든 질문에 답하였습니다. 사용자 정보를 바탕으로 컨텐츠를 추천합니다]")
        print("==========================================================================")
        a_imsi=1
        for i in list4_data:
            print(f"{a_imsi}. {sdata[1][i]} (감독: {sdata[2][i]})")
            a_imsi+=1
        print("==========================================================================")
        print("[원하는 컨텐츠 번호를 선택하시면 더 자세한 정보를 열람합니다]\n")

        a_imsi=1
        while a_imsi==1: 
            con=int(input("[컨텐츠 번호를 입력해주세요. (컨텐츠 추천 서비스를 종료하려면 '0'을 입력해주세요)] => "))

            if con==0:
                a_imsi-=1

            elif con > len(list4_data):
                print("\n[목록에 없는 번호입니다. 다시 확인해 주시기 바랍니다]")

            else:
                i=list4_data[con-1]
                print(f"""\n[선택하신 컨텐츠의 정보입니다]  
■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■                  

평점: {sd[i]} ({star_count[i]})
제목: {sdata[1][i]} (감독: {sdata[2][i]})
개봉 년도: {sdata[6][i]}
출연진: {sdata[3][i]}

구분: {sdata[11][i]} (개봉: {sdata[6][i]}년)
장르: {sdata[9][i]}

국가: {sdata[4][i]}
관람등급: {sdata[7][i]}
러닝 타임(시즌): {sdata[8][i]}

작품 설명: {sdata[10][i]} 

■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■■ 
[다른 컨텐츠의 정보를 확인하고 싶다면 추가로 컨텐츠를 선택하실 수 있습니다]\n""")
        
        print("[컨텐츠 추천 서비스를 종료합니다]")


    #############################################################
    #종료 (3)
    elif service=='3' or service=='종료':
        print("\n[넷플릭스 컨텐츠 룸((Netflix Contents Room)을 이용해주셔서 감사합니다]")
        print("[서비스를 종료합니다]")
        break
        
    #############################################################
    #잘못된 입력
    else:
        print("['검색' 혹은 '컨텐츠 추천' 중 하나를 선택해주세요]")
        continue
```
