```python
import keyboard, time, random
import numpy as np
from IPython.display import clear_output 
from datetime import datetime

#========================================================================

def makein(mlist, emo, n):    ####별, 장애물 랜덤 생성####
    global listblock
    
    while len(mlist)<n:
        sa=np.random.randint(8,size=1)   #세로
        sb=np.random.randint(10,size=1)   #가로
        if [sa[0],sb[0]] not in mlist and [sa[0],sb[0]]!=[7,9] and [sa[0],sb[0]]!=[0,0]:
            mlist.append([sa[0],sb[0]])
    for i,k in mlist:
        listblock[i][k]=emo
        
#========================================================================

def keyboardin(s1,s2):
    global a
    global listblock
    global snake

    try:
        sn_head=listblock[s1][s2]
        
        if sn_head!='::':
            snake.append([s1,s2])
            listblock[s1][s2]='■'  #머리 추가
            
    except:
        pass
    
    else:
        if sn_head=='  ':
            now2=datetime.now()
            result=now2-now1
            
            print("[Game Clear!]")
            print(f"▶Score: {len(snake)-1}/10")
            print(f"▶ClearTime: {result.seconds//60} min {result.seconds%60} second")
            a+=1   #겜 클리어
            
        elif sn_head == '□':
            listblock[snake[0][0]][snake[0][1]]='▩'  #끝에 거(지나온 길) 삭제
            del(snake[0])
                
        elif sn_head == '▩' or sn_head == '■':
            listblock[snake[0][0]][snake[0][1]]='▩'  #끝에거(지나온 길) 삭제
            del(snake[0])
            
            clear_output(wait=True)
            print("\n")   
            for j in listblock:
                print(f"{''.join(j)}")
            print("\n") 
            print("[Game Over]")
            
            a+=1  #게임 오버
        
        else:
            pass

#========================================================================

def load():
    global now1
    global star, bk, snake, listblock
    
    listblock=[['■','□','□','□','□','□','□','□','□','□'],['□','□','□','□','□','□','□','□','□','□'],['□','□','□','□','□','□','□','□','□','□'],['□','□','□','□','□','□','□','□','□','□'],['□','□','□','□','□','□','□','□','□','□'],['□','□','□','□','□','□','□','□','□','□'],['□','□','□','□','□','□','□','□','□','□'],['□','□','□','□','□','□','□','□','□','  ','→']]
    #[가로 10칸][세로 8칸]  listblock[7][9]
    star,bk,snake=[],[],[[0,0]]
    makein(bk,'::',20)
    makein(star,'☆',9)
    now1=datetime.now()

#========================================================================

a=1 
load() 
while a==1:
    sn1=snake[-1][0]    #세로
    sn2=snake[-1][1]    #가로
            
    ####게임 화면 출력####
    clear_output(wait=True) 
    print("\n") 
    for j in listblock:
        print(f"{''.join(j)}")
    print("\n") 
    time.sleep(0.2)
        
    if keyboard.is_pressed('esc'):
        time.sleep(0.1)
        sel=input('[게임을 종료하시겠습니까? (Y/N)] ')
        if sel =='y' or sel=='Y' or sel=='Yes' or sel=='yes':
            break
                
    elif keyboard.is_pressed('up'):
        keyboardin(sn1-1,sn2)
        
    elif keyboard.is_pressed('down'):
        keyboardin(sn1+1,sn2)
        
    elif keyboard.is_pressed('right'):
        keyboardin(sn1,sn2+1)
        
    elif keyboard.is_pressed('left'):
        keyboardin(sn1,sn2-1)
    
    elif keyboard.is_pressed('enter'):
        clear_output(wait=True)
        print('\n[게임을 새로 불러옵니다.]')      #다시 시작(게임판 재설정)
        
        time.sleep(0.5)
        load()
```
